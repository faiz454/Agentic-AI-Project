```python
%%writefile app.py
import streamlit as st
import datetime
import pandas as pd

# 1. Agentic Logic Engine
class StudyAgent:
    @staticmethod
    def get_ai_advice(message):
        message = message.lower()
        if "stress" in message or "worried" in message:
            return "AI Agent: I've noticed your concern. I am reducing today's workload by 20% to help you focus better. Take a 10-minute break first. üßò‚Äç‚ôÇÔ∏è"
        elif "hard" in message or "difficult" in message:
            return "AI Agent: I will break down this subject into smaller 30-minute 'sprints' to make it easier for you. ‚ö°"
        else:
            return "AI Agent: Keep going! Based on your exam date, you are exactly where you need to be. üéØ"

# 2. Professional Page Config
st.set_page_config(page_title="Agentic AI Study Hub", layout="wide")
st.title("ü§ñ Agentic AI Personal Study Planner")
st.caption("Innovation by: MUHAMMED FAIZ C K | Reg No: 2411021240042")

# 3. Sidebar - The "Observer" (Section 5 of Idea)
st.sidebar.header("üìä Student Performance Data")
subject = st.sidebar.selectbox("Current Subject", ["Mathematics", "Physics", "Computer Science"])
exam_date = st.sidebar.date_input("Exam Date", datetime.date(2025, 12, 30))
chapters_total = st.sidebar.number_input("Total Chapters", value=10)
chapters_done = st.sidebar.number_input("Completed Chapters", value=2)

# 4. Main Dashboard - "Active Management" (Section 6 of Idea)
col1, col2 = st.columns([2, 1])

with col1:
    st.subheader("üí¨ AI Study Chatbot")
    if "messages" not in st.session_state:
        st.session_state.messages = []

    # Display chat history
    for msg in st.session_state.messages:
        with st.chat_message(msg["role"]):
            st.markdown(msg["content"])

    # Chat input
    if prompt := st.chat_input("Tell the AI how your studies are going..."):
        st.session_state.messages.append({"role": "user", "content": prompt})
        with st.chat_message("user"):
            st.markdown(prompt)

        # Agentic Response
        response = StudyAgent.get_ai_advice(prompt)
        st.session_state.messages.append({"role": "assistant", "content": response})
        with st.chat_message("assistant"):
            st.markdown(response)

with col2:
    st.subheader("üìà Live Progress")
    progress = chapters_done / chapters_total
    st.progress(progress)
    st.metric("Completion", f"{int(progress*100)}%", "+5% from yesterday")

    st.subheader("ü§ñ AI Decision")
    days_left = (exam_date - datetime.date.today()).days
    if days_left < 7:
        st.error(f"Urgent: {days_left} days left! AI has locked your schedule to 'Intensive Mode'.")
    else:
        st.success("Status: Balanced Mode. AI is maintaining a healthy study-life balance.")
```

```python
import datetime

class AgenticStudyPlanner:
    def __init__(self, student_name, subjects):
        self.student_name = student_name
        self.subjects = subjects  # Dictionary: {subject: difficulty_score}
        self.schedule = {}

    def create_initial_plan(self):
        """Creates a daily study plan based on subjects [cite: 15]"""
        print(f"--- Initializing Plan for {self.student_name} ---")
        for subject, difficulty in self.subjects.items():
            # More difficult subjects get more hours
            hours = 1 if difficulty < 5 else 2
            self.schedule[subject] = {"hours": hours, "status": "Pending"}
        return self.schedule

    def observe_and_adjust(self, completed_tasks):
        """
        Agentic AI: Observes performance and adjusts the plan automatically [cite: 19, 22]
        """
        print("\n--- Agentic AI is analyzing performance... ---")

        adjusted_tasks = []
        for subject, status in completed_tasks.items():
            if status == "Missed":
                # The AI decides to increase focus on missed subjects [cite: 21]
                print(f"Action: {subject} was missed. Increasing priority for tomorrow.")
                self.schedule[subject]["hours"] += 0.5
                adjusted_tasks.append(subject)
            else:
                print(f"Action: {subject} completed. Well done!")

        if not adjusted_tasks:
            return "Progress is on track! Keep it up. "
        else:
            return f"Schedule adjusted. Added focus to: {', '.join(adjusted_tasks)} [cite: 24]"

# --- Running the Simulation ---

# 1. Input: Name and Subjects [cite: 2, 14]
my_subjects = {"Mathematics": 8, "History": 4, "Physics": 7}
ai_agent = AgenticStudyPlanner("Muhammed Faiz C K", my_subjects)

# 2. AI creates the daily and weekly plan [cite: 15]
current_plan = ai_agent.create_initial_plan()
print("Current Schedule:", current_plan)

# 3. AI observes student behavior (Simulated feedback) [cite: 20]
student_feedback = {
    "Mathematics": "Completed",
    "History": "Completed",
    "Physics": "Missed"
}

# 4. AI makes decisions and takes action [cite: 19, 22]
update_message = ai_agent.observe_and_adjust(student_feedback)
print(f"AI Message: {update_message}")
```

```python
import random

class AgenticStudyPlanner:
    def __init__(self, student_name, subjects):
        self.student_name = student_name
        self.subjects = subjects
        self.schedule = {}
        # New feature: Tracking history for progress reports
        self.history = {"Completed": 0, "Missed": 0}

    def create_initial_plan(self):
        print(f"--- Initializing Plan for {self.student_name} ---")
        for subject, difficulty in self.subjects.items():
            hours = 1 if difficulty < 5 else 2
            self.schedule[subject] = {"hours": hours, "status": "Pending"}
        return self.schedule

    def get_motivation(self, success_rate):
        """Generates motivation messages based on behavior """
        if success_rate > 0.7:
            return "üî• Outstanding work! You are mastering your syllabus."
        elif success_rate > 0:
            return "üí™ Good effort today. Let's close the gaps tomorrow!"
        else:
            return "üõãÔ∏è Take a deep breath. Tomorrow is a fresh start to reach your goals."

    def observe_and_adjust(self, completed_tasks):
        print(f"\n--- {self.student_name}'s Agentic AI Analysis ---")

        adjusted_tasks = []
        for subject, status in completed_tasks.items():
            if status == "Missed":
                self.schedule[subject]["hours"] += 0.5
                self.history["Missed"] += 1
                adjusted_tasks.append(subject)
            else:
                self.history["Completed"] += 1

        # Calculate success for the Progress Report
        total = self.history["Completed"] + self.history["Missed"]
        success_rate = self.history["Completed"] / total

        print(f"üìä PROGRESS REPORT: You completed {int(success_rate*100)}% of your tasks.")
        print(f"BOTTLE-NECK ALERT: {', '.join(adjusted_tasks) if adjusted_tasks else 'None'}")
        print(f"AI MOTIVATION: {self.get_motivation(success_rate)}")

# --- Test the New Features ---
my_subjects = {"Mathematics": 8, "History": 4, "Physics": 7}
ai_agent = AgenticStudyPlanner("MUHAMMED FAIZ C K", my_subjects)
ai_agent.create_initial_plan()

# Simulating the student behavior again
student_feedback = {"Mathematics": "Completed", "History": "Completed", "Physics": "Missed"}
ai_agent.observe_and_adjust(student_feedback)
```

```python
import datetime

class AgenticStudyPlanner:
    def __init__(self, student_name, syllabus_info):
        self.student_name = student_name
        self.syllabus = syllabus_info  # {Subject: {"difficulty": 1-10, "exam_date": datetime}}
        self.schedule = {}

    def analyze_urgency(self):
        """
        The Agentic Aspect: AI calculates which subject needs attention
        based on how many days are left until the exam.
        """
        print(f"\n--- AI Deadline Analysis for {self.student_name} ---")
        today = datetime.date.today()

        for subject, info in self.syllabus.items():
            days_left = (info["exam_date"] - today).days

            # Logic: If exam is in less than 7 days, increase difficulty/focus weight
            if days_left <= 7:
                print(f"‚ö†Ô∏è URGENT: {subject} exam is in {days_left} days! Adjusting focus.")
                info["difficulty"] += 2  # AI takes action to prioritize this
            else:
                print(f"üìÖ {subject} exam is in {days_left} days. Schedule remains steady.")

    def create_priority_plan(self):
        self.analyze_urgency()
        print("\n--- Generating Priority Schedule ---")
        for subject, info in self.syllabus.items():
            # AI decides hours based on difficulty and urgency
            hours = 3 if info["difficulty"] >= 8 else 1.5
            self.schedule[subject] = f"{hours} hours"
            print(f"Plan: Study {subject} for {hours} hours.")

# --- Running the Advanced Simulation ---

# 1. Input: Syllabus and Exam Dates (Year, Month, Day)
# Let's pretend today is Dec 17, 2025.
syllabus_data = {
    "Mathematics": {"difficulty": 8, "exam_date": datetime.date(2025, 12, 20)}, # Very soon
    "History": {"difficulty": 4, "exam_date": datetime.date(2026, 1, 15)}      # Far away
}

ai_agent = AgenticStudyPlanner("MUHAMMED FAIZ C K", syllabus_data)

# 2. AI calculates urgency and creates the plan
ai_agent.create_priority_plan()
```

```python
import datetime

class AgenticStudyPlanner:
    def __init__(self, student_name, syllabus_info):
        self.student_name = student_name
        # Format: {Subject: {"difficulty": 1-10, "exam_date": date, "total_chapters": int, "chapters_done": int}}
        self.syllabus = syllabus_info
        self.performance_history = []

    def generate_progress_report(self):
        """AI generates progress reports as described in the proposal."""
        print(f"\n--- üìä PROGRESS REPORT FOR {self.student_name} ---")
        for subject, info in self.syllabus.items():
            completion = (info["chapters_done"] / info["total_chapters"]) * 100
            print(f"{subject}: {completion:.1f}% complete ({info['chapters_done']}/{info['total_chapters']} chapters)")

    def auto_adjust_schedule(self, subject_worked_on, chapters_completed):
        """
        The Core Innovation: Agentic AI decides what needs more focus
        and changes the study plan without manual input.
        """
        print(f"\n--- ü§ñ Agentic AI Observing: {subject_worked_on} ---")

        # Update progress data
        self.syllabus[subject_worked_on]["chapters_done"] += chapters_completed
        remaining = self.syllabus[subject_worked_on]["total_chapters"] - self.syllabus[subject_worked_on]["chapters_done"]

        if chapters_completed == 0:
            # Action: Automatically adjusting the plan if the student misses a task
            print(f"‚ö†Ô∏è Action: {subject_worked_on} was missed. AI is increasing focus priority.")
            self.syllabus[subject_worked_on]["difficulty"] += 1
        else:
            # Motivation: Giving reminders and positive feedback
            print(f"‚úÖ Action: {chapters_completed} chapters completed. AI updated your goals.")

        if remaining <= 0:
            print(f"üéâ Goal Achieved: Syllabus for {subject_worked_on} is complete!")

# --- 1. SETUP: Input Student and Syllabus Details ---
# As per the idea: Student enters subjects, syllabus details, and exam dates.
syllabus_data = {
    "Mathematics": {
        "difficulty": 8,
        "exam_date": datetime.date(2025, 12, 25),
        "total_chapters": 10,
        "chapters_done": 2
    },
    "Physics": {
        "difficulty": 7,
        "exam_date": datetime.date(2025, 12, 30),
        "total_chapters": 8,
        "chapters_done": 1
    }
}

# --- 2. EXECUTION: AI starts monitoring and planning ---
ai_agent = AgenticStudyPlanner("MUHAMMED FAIZ C K", syllabus_data)

# Show initial progress
ai_agent.generate_progress_report()

# Simulate a missed task (Mathematics) and a completed task (Physics)
ai_agent.auto_adjust_schedule("Mathematics", 0)
ai_agent.auto_adjust_schedule("Physics", 2)

# Final adjusted progress report
ai_agent.generate_progress_report()
```

```python
import streamlit as st
import datetime
import pandas as pd

# --- Agentic Logic Class ---
class AgenticEngine:
    @staticmethod
    def calculate_urgency(exam_date):
        days_left = (exam_date - datetime.date.today()).days
        return "High" if days_left <= 7 else "Normal"

    @staticmethod
    def get_motivation(progress):
        if progress >= 80: return "üåü You're almost at the finish line!"
        if progress >= 50: return "üìà Great momentum! Keep pushing."
        return "üöÄ Every chapter counts. Start today!"

# --- Web App UI ---
st.set_page_config(page_title="Agentic AI Study Planner", layout="wide")

st.title("ü§ñ Agentic AI Personal Study Planner")
st.markdown(f"**Student:** MUHAMMED FAIZ C K | **ID:** 2411021240042") # [cite: 2, 3]

# Sidebar for Input [cite: 14]
st.sidebar.header("üìã Input Syllabus & Exams")
subject = st.sidebar.text_input("Subject Name")
total_ch = st.sidebar.number_input("Total Chapters", min_value=1, value=10)
done_ch = st.sidebar.number_input("Chapters Completed", min_value=0, value=0)
exam_dt = st.sidebar.date_input("Exam Date", datetime.date(2025, 12, 30))

if st.sidebar.button("Add/Update Subject"):
    st.success(f"AI has integrated {subject} into your agentic plan.")

# Main Dashboard
col1, col2 = st.columns(2)

with col1:
    st.subheader("üìÖ Smart Study Schedule")
    # Simulation of Agentic Adjustment [cite: 12, 16, 22]
    urgency = AgenticEngine.calculate_urgency(exam_dt)

    status_color = "red" if urgency == "High" else "green"
    st.markdown(f"**Priority Status:** :{status_color}[{urgency} Priority]")

    st.info(f"**AI Decision:** Based on your exam date ({exam_dt}), the Agentic AI has "
            f"{'increased' if urgency == 'High' else 'optimized'} your daily study hours.") # [cite: 19, 21]

with col2:
    st.subheader("üìä Progress Analytics")
    progress_perc = (done_ch / total_ch)
    st.progress(progress_perc)
    st.write(f"Completion: {int(progress_perc*100)}%")
    st.write(f"**AI Message:** {AgenticEngine.get_motivation(progress_perc*100)}") #

# Performance Monitoring Table [cite: 20, 24]
st.divider()
st.subheader("üìù Activity Observer")
data = {
    "Subject": ["Mathematics", "Physics", "History"],
    "AI Adjustment": ["+1.5 hrs (Missed Task)", "Optimized", "Steady"],
    "Status": ["Urgent", "On Track", "Ahead"]
}
st.table(pd.DataFrame(data))
```
